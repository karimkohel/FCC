```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Dataframe preview

```python
df = pd.read_csv("../dataset/dataset.csv")
df.drop(columns = df.columns[0], axis=1, inplace=True)
df
```

# All car images count

```python
fig, ax = plt.subplots(figsize=(25, 9))
plot = sns.barplot(x=df['car'], y=df['train'], ax=ax).set(title="All")
```

# Lowest image count cars

```python
sortedDF = df.sort_values(by='train')
fig2, ax2 = plt.subplots(figsize=(25, 9))
sortedDFPlot = sns.barplot(x=sortedDF['car'][:15], y=sortedDF['train'][:15], ax=ax2).set(title="Sorted Lowest")
```

# Testing

```python
testDf = pd.read_csv("../docs/runs/run2score.csv")
testDf.drop(columns = testDf.columns[0], axis=1, inplace=True)
fig3, ax3 = plt.subplots(figsize=(25, 9))
testPlot = sns.barplot(x=testDf['car'], y=testDf['score'], ax=ax3).set(title="Test Scores")
```

# Filtering results

```python
badCarsDf = testDf[testDf['score'] < 10]
fig4, ax4 = plt.subplots(figsize=(20, 9))
badCarsDfPlot = sns.barplot(x=badCarsDf['car'], y=badCarsDf['score'], ax=ax4).set(title="Worst scores")
```
